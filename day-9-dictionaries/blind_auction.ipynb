```python
# Importing the 'clear' function from the 'replit' module (commented out as it is not used)
# from replit import clear

# Displaying a welcome message
print('Welcome to the Secret Auction Program')

# Initializing a flag to control the loop
more_people = True

# Initializing an empty list to store people and their bids
list_people = []

# Function to add a person and their bid to the list
def add_people(name, price):
    new_person = {}
    new_person[name] = price
    list_people.append(new_person)

# Loop to collect input for people and their bids
while more_people:
    # Collecting name and price from the user
    name = input('What is your name? ')
    price = int(input('What is your price? '))
    
    # Adding the person and their bid to the list
    add_people(name, price)

    # Asking the user if there's another person
    cont = input('Do you have another person? (yes/no)').lower()

    # Checking the user's response to determine whether to continue the loop
    if cont == 'no':
        more_people = False

# Initializing an empty dictionary to store bids
new_dict = {}

# Looping through each person in the list
for person in list_people:
    # Since each person is a dictionary with one key-value pair, use items() to access it
    for key, value in person.items():
        # Adding the bid to the dictionary
        new_dict[key] = value

# Function to find the highest bid and the corresponding winner
def find_highest(bid_record):
    highest_bid = 0
    winner = ''
    # Looping through each bidder in the dictionary
    for bidder in bid_record:
        # Retrieving the bid amount for the current bidder
        bid_amount = bid_record[bidder]
        # Checking if the current bid is higher than the previous highest bid
        if bid_amount > highest_bid:
            highest_bid = bid_amount
            winner = bidder
    # Displaying the winner and their bid
    print(f'The winner is {winner} and their bid was {highest_bid}')

# Calling the function to find and display the highest bid and winner
find_highest(new_dict)
```

```python
def welcome_message():
    print('Welcome to the Secret Auction Program')

def add_person(name, price, people_list):
    new_person = {name: price}
    people_list.append(new_person)

def get_input():
    name = input('What is your name? ')
    price = int(input('What is your price? '))
    return name, price

def continue_auction():
    return input('Do you have another person? (yes/no)').lower() == 'yes'

def create_dictionary(people_list):
    new_dict = {}
    for item in people_list:
        key, value = item
        new_dict[key] = value
    return new_dict

def find_highest_bid(bid_record):
    highest_bid = 0
    winner = ''
    for bidder, bid_amount in bid_record.items():
        if bid_amount > highest_bid:
            highest_bid = bid_amount
            winner = bidder
    print(f'The winner is {winner} and their bid was {highest_bid}')

def main():
    welcome_message()
    
    more_people = True
    people_list = []

    while more_people:
        name, price = get_input()
        add_person(name, price, people_list)

        more_people = continue_auction()

    new_dict = create_dictionary(people_list)
    find_highest_bid(new_dict)

if __name__ == "__main__":
    main()
```
