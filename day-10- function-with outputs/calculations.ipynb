```python
#calculations

#Add
def addition(num1, num2):
    return num1+num2
#Subtract
def subtract(num1, num2):
    return num1-num2
#divide
def divide(num1, num2):
    if num2==0:
        return
    return num1/num2
#multiply
def multiply(num1, num2):
    return num1*num2

calc_dict={
    '+': addition,
    '-': subtract,
    '/':divide,
    '*':multiply
}
# num1=int(input('What is the first number? '))
# num2=int(input('What is the second number? '))

# for key in calc_dict:
#     print(f'{key}\n')
# operation_user= input('Pick an operation from above')

# calc_funct= calc_dict[operation_user]
# answer=calc_funct(num1, num2)

def calculator():
    num1= int(input('What is the first number '))
    for symbol in calc_dict:
        print(symbol)

    continue_or= True
    while continue_or:
        operation= input('What opeation do you pick? ')

        num2= int('What is your second number?')

        calc_funct= calc_dict[operation]
        answer= calc_funct(num1, num2)

        print(f' {num1} {operation} {num2} = {answer}')

        if input(' y for  or n')=='y':
            num1= answer
        else: 
            continue_or=False
            calculator()

calculator()
```
